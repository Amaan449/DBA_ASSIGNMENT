```python
# Define a mapping from common zone spellings to standardized zone names
zone_map = {
    "north": "North",
    "south": "South",
    "east": "East",
    "west": "West",
    "central": "Central",
    "ctr": "Central",
    "riverside": "Riverside",
    "airpor": "Airport",
    "airport": "Airport"
}

# Function to clean and standardize zone values using the zone_map
def clean_zone(x):
    if pd.isna(x):
        return np.nan
    x = str(x).strip()
    key = x.lower()
    return zone_map.get(key, x.title())
# Apply the clean_zone function to standardize zone values in relevant columns
if "home_zone" in customers.columns:
    customers["home_zone"] = customers["home_zone"].apply(clean_zone)

if "pickup_zone" in orders.columns:
    orders["pickup_zone"] = orders["pickup_zone"].apply(clean_zone)

if "dropoff_zone" in orders.columns:
    orders["dropoff_zone"] = orders["dropoff_zone"].apply(clean_zone)

if "base_zone" in drivers.columns:
    drivers["base_zone"] = drivers["base_zone"].apply(clean_zone)

if "zone" in hubs.columns:
    hubs["zone"] = hubs["zone"].apply(clean_zone)

if "zone_context" in app_events.columns:
    app_events["zone_context"] = app_events["zone_context"].apply(clean_zone)


# Show unique non‑missing values for each zone column after cleaning
for col in ["home_zone", "pickup_zone", "dropoff_zone", "base_zone", "zone", "zone_context"]:
    print(f"\n--- {col} AFTER CLEANING ---")
    for name, df in {
        "customers": customers,
        "orders": orders,
        "drivers": drivers,
        "hubs": hubs,
        "app_events": app_events
    }.items():
        if col in df.columns:
            print(name, df[col].dropna().unique())
```

```python
# Convert datetime columns
date_cols = {
    "customers": ["signup_date"],
    "orders": ["order_created_at"],
    "deliveries": ["dispatch_time", "delivery_completed_at"],
    "complaints": ["created_at"],
    "incidents": ["reported_at"],
    "app_events": ["event_timestamp"]
}

dfs = {
    "customers": customers,
    "orders": orders,
    "deliveries": deliveries,
    "complaints": complaints,
    "incidents": incidents,
    "app_events": app_events
}

for name, cols in date_cols.items():
    df = dfs[name]
    for col in cols:
        if col in df.columns:
            df[col] = pd.to_datetime(df[col], errors="coerce")

# Convert numeric columns
numeric_conversions = {
    "customers": ["loyalty_score", "app_engagement_score"],
    "orders": ["promised_window_hours", "order_value", "special_handling_flag"],
    "deliveries": ["route_distance_km", "manual_route_override_count",
                   "proof_of_completion_missing", "customer_rating_post_delivery",
                   "fuel_or_charge_cost"],
    "drivers": ["years_experience", "training_score", "driver_rating", "active_flag"],
    "vehicles": ["battery_health_pct", "maintenance_status"],
    "complaints": ["compensation_amount"],
    "incidents": ["resolved_hours"],
    "app_events": ["api_latency_ms", "success_flag"]
}

dfs_num = {
    "customers": customers,
    "orders": orders,
    "deliveries": deliveries,
    "drivers": drivers,
    "vehicles": vehicles,
    "complaints": complaints,
    "incidents": incidents,
    "app_events": app_events
}

for name, cols in numeric_conversions.items():
    df = dfs_num[name]
    for col in cols:
        if col in df.columns:
            df[col] = pd.to_numeric(df[col], errors="coerce")

print("Data type conversion finished.")
```

```python
# Merging attach delivery, customer, hub, driver, and vehicle data

# Enrich orders with delivery information by merging on order_id
orders_full = orders.merge(
    deliveries,
    on="order_id",
    how="left",
    suffixes=("_order", "_del")
)

# Enrich the orders–deliveries table with selected customer features by merging on customer_id
orders_full = orders_full.merge(
    customers[[
        "customer_id",
        "age",
        "home_zone",
        "customer_type",
        "loyalty_score",
        "app_engagement_score",
        "preferred_channel",
        "account_status"
    ]],
    on="customer_id",
    how="left"
)

# Add hub details (hub_name, zone, hub_type, capacity_score) to orders_full by merging on hub_id
orders_full = orders_full.merge(
    hubs[["hub_id", "hub_name", "zone", "hub_type", "capacity_score"]],
    on="hub_id",
    how="left"
)

# Enrich orders with driver attributes by merging on driver_id
orders_full = orders_full.merge(
    drivers[[
        "driver_id",
        "base_zone",
        "employment_type",
        "years_experience",
        "training_score",
        "driver_rating",
        "shift_preference",
        "active_flag"
    ]],
    on="driver_id",
    how="left"
)

# Add vehicle details (e.g., vehicle_type, battery_health_pct, maintenance_status) to orders by merging on vehicle_id
orders_full = orders_full.merge(
    vehicles,
    on="vehicle_id",
    how="left"
)

# OUTPUT ONLY AFTER ALL JOINS ARE COMPLETED
print(orders_full.shape)
orders_full.head()
```

```python

# Aggregate complaint data by order_id to create summary features for each order
complaints_by_order = (
    complaints
    .groupby("order_id")
    .agg(
        complaint_count=("complaint_id", "count"),
        last_complaint_date=("created_at", "max"),
        any_high_severity=("severity", lambda s: (s == "High").any()),
        avg_compensation=("compensation_amount", "mean")
    )
    .reset_index()
)

complaints_by_order["any_high_severity"] = complaints_by_order["any_high_severity"].astype(int)
complaints_by_order.head()


# Attach aggregated complaint features to orders by merging with complaints_by_order on order_id
orders_full = orders_full.merge(
    complaints_by_order,
    on="order_id",
    how="left"
)

# Aggregate app event data by order_id to create summary features for each order
app_events_by_order = (
    app_events.dropna(subset=["order_id"])
    .groupby("order_id")
    .agg(
        app_event_count=("event_id", "count"),
        last_app_event=("event_timestamp", "max"),
        unique_event_types=("event_type", "nunique"),
        avg_api_latency=("api_latency_ms", "mean")
    )
    .reset_index()
)

app_events_by_order["avg_api_latency"] = app_events_by_order["avg_api_latency"].round(1)

# Attach aggregated app event features to orders_full
orders_full = orders_full.merge(
    app_events_by_order,
    on="order_id",
    how="left"
)

# OUTPUT ONLY AFTER ALL JOINS ARE COMPLETED
print(orders_full.shape)
orders_full.head()
```

```python
# Finalize the orders_full table by deduplicating on order_id and handling missing values for derived features
orders_full = orders_full.drop_duplicates(subset=["order_id"], keep="first")

orders_full["complaint_count"] = orders_full["complaint_count"].fillna(0)
orders_full["app_event_count"] = orders_full["app_event_count"].fillna(0)
orders_full["any_high_severity"] = orders_full["any_high_severity"].fillna(0)
orders_full["avg_compensation"] = orders_full["avg_compensation"].fillna(0)

print("Rows:", orders_full.shape[0])
print("Columns:", orders_full.shape[1])
print("Duplicate order_ids:", orders_full["order_id"].duplicated().sum())
print(orders_full[["order_id", "delivery_id", "complaint_count", "app_event_count"]].head(10))
print(orders_full[["complaint_count", "app_event_count", "any_high_severity", "avg_compensation"]].isna().sum())
```

```python
%%R
# Query 1: Delay performance by pickup zone
zone_delay <- sqldf("
  SELECT pickup_zone,
         COUNT(*) AS order_count,
         ROUND(AVG(delay_hours), 2) AS avg_delay_hrs,
         ROUND(SUM(CASE WHEN is_delayed=1 THEN 1.0 ELSE 0 END) / COUNT(*), 3) AS failure_rate
  FROM df
  WHERE delay_hours IS NOT NULL
  GROUP BY pickup_zone
  ORDER BY avg_delay_hrs DESC
")
print("--- Q1: Delay by zone ---")
print(zone_delay)
```

```python
%%R
# Query 2: Complaint intensity by service type
complaint_svc <- sqldf("
  SELECT service_type,
         COUNT(*) AS order_count,
         ROUND(AVG(complaint_count), 3) AS avg_complaints,
         ROUND(AVG(avg_compensation), 2) AS avg_comp_paid,
         ROUND(AVG(is_delayed), 3) AS delay_rate
  FROM df
  GROUP BY service_type
  ORDER BY delay_rate DESC
")
print("--- Q2: Complaints by service type ---")
print(complaint_svc)
```

```python
%%R
# Query 3: High-override routes
override_routes <- sqldf("
  SELECT pickup_zone || ' → ' || dropoff_zone AS route,
         COUNT(*) AS orders,
         ROUND(AVG(manual_route_override_count), 2) AS avg_overrides,
         ROUND(AVG(delay_hours), 2) AS avg_delay,
         ROUND(AVG(fuel_or_charge_cost), 2) AS avg_cost
  FROM df
  WHERE manual_route_override_count > 0
  GROUP BY route
  ORDER BY avg_overrides DESC
  LIMIT 10
")
print("--- Q3: High-override routes ---")
print(override_routes)
```

```python
%%R
# Query 4: Hub-level performance
hub_perf <- sqldf("
  SELECT hub_name,
         COUNT(*) AS orders_dispatched,
         ROUND(AVG(is_delayed), 3) AS delay_rate,
         ROUND(AVG(complaint_count), 3) AS avg_complaints,
         ROUND(AVG(manual_route_override_count), 2) AS avg_overrides,
         ROUND(AVG(order_value), 2) AS avg_order_value,
         ROUND(SUM(fuel_or_charge_cost), 2) AS total_fuel_cost
  FROM df
  GROUP BY hub_name
  ORDER BY orders_dispatched DESC
")
print("--- Q4: Hub-level performance ---")
print(hub_perf)
```

```python
%%R
# Query 5: Driver risk profile (top 15 high-override drivers)
driver_risk <- sqldf("
  SELECT driver_id,
         employment_type,
         ROUND(AVG(years_experience), 1) AS avg_experience,
         ROUND(AVG(driver_rating), 2) AS avg_rating,
         ROUND(AVG(training_score), 2) AS avg_training,
         COUNT(*) AS orders_handled,
         ROUND(AVG(manual_route_override_count), 2) AS avg_overrides,
         ROUND(AVG(is_delayed), 3) AS delay_rate
  FROM df
  WHERE driver_id IS NOT NULL
  GROUP BY driver_id, employment_type
  HAVING COUNT(*) >= 3
  ORDER BY avg_overrides DESC
  LIMIT 15
")
print("--- Q5: Top 15 high-override drivers ---")
print(driver_risk)
```

```python
%%R
# Query 6: Repeat-complaint customers
repeat_customers <- sqldf("
  SELECT customer_id,
         customer_type,
         home_zone,
         COUNT(*) AS orders,
         ROUND(SUM(complaint_count), 0) AS total_complaints,
         ROUND(AVG(avg_compensation), 2) AS avg_comp_received,
         ROUND(AVG(loyalty_score), 2) AS loyalty_score,
         ROUND(AVG(is_delayed), 3) AS delay_rate
  FROM df
  GROUP BY customer_id, customer_type, home_zone
  HAVING SUM(complaint_count) >= 2
  ORDER BY total_complaints DESC
  LIMIT 15
")
print("--- Q6: Repeat-complaint customers (top 15) ---")
print(repeat_customers)
```

```python
%%R
# Query 7: Vehicle maintenance vs delay
vehicle_maint <- sqldf("
  SELECT vehicle_type,
         CASE WHEN maintenance_status = 'InRepair' THEN 'Under Maintenance'
              ELSE 'Operational' END AS maint_status,
         COUNT(*) AS orders,
         ROUND(AVG(delay_hours), 2) AS avg_delay,
         ROUND(AVG(battery_health_pct), 1) AS avg_battery,
         ROUND(AVG(is_delayed), 3) AS delay_rate,
         ROUND(AVG(fuel_or_charge_cost), 2) AS avg_fuel_cost
  FROM df
  WHERE vehicle_type IS NOT NULL
  GROUP BY vehicle_type, maint_status
  ORDER BY vehicle_type, maint_status
")
print("--- Q7: Vehicle maintenance vs delay ---")
print(vehicle_maint)
```

```python
# --- Q8: Monthly trend ---

# Create yr_month from order_created_at
orders["yr_month"] = pd.to_datetime(orders["order_created_at"]).dt.to_period("M")

# Merge orders with deliveries to get delay information
orders_with_delivery = orders.merge(
    deliveries[["order_id", "dispatch_time", "delivery_completed_at"]],
    on="order_id",
    how="left"
)

# Calculate delay (if delivery_completed_at exists)
orders_with_delivery["delivery_completed_at"] = pd.to_datetime(orders_with_delivery["delivery_completed_at"])
orders_with_delivery["order_created_at_dt"] = pd.to_datetime(orders_with_delivery["order_created_at"])
orders_with_delivery["promised_deadline"] = orders_with_delivery["order_created_at_dt"] + pd.to_timedelta(orders_with_delivery["promised_window_hours"], unit="h")
orders_with_delivery["delay_hours"] = (orders_with_delivery["delivery_completed_at"] - orders_with_delivery["promised_deadline"]).dt.total_seconds() / 3600

# Count complaints per order
complaints_per_order = complaints.groupby("order_id").size().reset_index(name="complaint_count")

monthly_trend = (
    orders_with_delivery.merge(complaints_per_order, on="order_id", how="left")
    .groupby("yr_month")
    .agg(
        orders=("order_id", "count"),
        delay_rate=("delay_hours", lambda x: (x > 0).mean()),  # % of orders with delay > 0
        avg_value=("order_value", "mean"),
        avg_complaints=("complaint_count", "mean")  # avg complaints per order
    )
    .reset_index()
)

# Format yr_month as string for display
monthly_trend["yr_month"] = monthly_trend["yr_month"].astype(str)

print(monthly_trend)
```

```python
%%R
# Query 9: INSERT, UPDATE, DELETE demonstrations
# INSERT equivalent: escalation log
escalation_log <- sqldf("
  SELECT order_id,
         'High delay + complaint' AS flag_reason,
         datetime('now') AS flagged_at
  FROM df
  WHERE delay_hours > 5 AND complaint_count >= 1
  LIMIT 10
")
print("--- Q9a: INSERT — escalation log sample ---")
print(escalation_log)

# UPDATE equivalent: reclassify borderline deliveries
df_update_demo <- df %>%
  mutate(delivery_status_revised = case_when(
    delivery_status == "OnTime" & delay_hours > 0.5 ~ "Borderline",
    TRUE ~ delivery_status
  ))
print("--- Q9b: UPDATE — borderline reclassification ---")
print(table(df_update_demo$delivery_status_revised, useNA = "ifany"))

# DELETE equivalent: keep only rows with non‑null delivery_id
df_clean <- sqldf("SELECT * FROM df WHERE delivery_id IS NOT NULL AND delivery_id != ''")
cat("--- Q9c: DELETE equivalent — rows with delivery_id ---\n")
cat("Rows kept:", nrow(df_clean), "of", nrow(df), "\n")
```

```python
%%R
# Mathematical and aggregate functions: overall summary
summary_stats <- sqldf("
  SELECT COUNT(*) AS total_orders,
         ROUND(AVG(order_value), 2) AS mean_order_value,
         ROUND(MIN(order_value), 2) AS min_order_value,
         ROUND(MAX(order_value), 2) AS max_order_value,
         ROUND(SUM(order_value), 2) AS total_revenue,
         ROUND(AVG(delay_hours), 2) AS mean_delay_hrs,
         ROUND(AVG(route_distance_km), 2) AS mean_route_km,
         ROUND(AVG(fuel_or_charge_cost), 2) AS mean_fuel_cost,
         ROUND(SUM(fuel_or_charge_cost), 2) AS total_fuel_cost,
         ROUND(AVG(complaint_count), 4) AS mean_complaints_per_order,
         ROUND(SUM(complaint_count), 0) AS total_complaints,
         ROUND(SUM(avg_compensation), 2) AS total_compensation_paid
  FROM df
")
print("--- Operational Summary Statistics ---")
print(t(summary_stats))

# Cost per km by vehicle type
cost_per_km <- sqldf("
  SELECT vehicle_type,
         COUNT(*) AS orders,
         ROUND(SUM(fuel_or_charge_cost), 2) AS total_cost,
         ROUND(SUM(route_distance_km), 2) AS total_km,
         ROUND(SUM(fuel_or_charge_cost)/SUM(route_distance_km), 3) AS cost_per_km
  FROM df
  WHERE route_distance_km > 0
  GROUP BY vehicle_type
  ORDER BY cost_per_km DESC
")
print("--- Cost per km by vehicle type ---")
print(cost_per_km)
```

##Statistical Validation

```python
%%R
# Filter for Central and South zones (non‑missing delay_hours)
central_delays <- df$delay_hours[df$pickup_zone == "Central" & !is.na(df$delay_hours)]
south_delays <- df$delay_hours[df$pickup_zone == "South" & !is.na(df$delay_hours)]

cat("Central zone delay hours (n=", length(central_delays), "): mean =", round(mean(central_delays),2), ", sd =", round(sd(central_delays),2), "\n")
cat("South zone delay hours (n=", length(south_delays), "): mean =", round(mean(south_delays),2), ", sd =", round(sd(south_delays),2), "\n")
```

```python
%%R
# Perform t-test assuming unequal variance (Welch's test)
t_test_result <- t.test(central_delays, south_delays, alternative = "two.sided", var.equal = FALSE)
print(t_test_result)

# Extract key values for reporting
cat("\n--- Summary for Report ---\n")
cat("t-statistic:", round(t_test_result$statistic, 2), "\n")
cat("p-value:", format(t_test_result$p.value, scientific = FALSE), "\n")
cat("Degrees of freedom:", round(t_test_result$parameter, 2), "\n")
cat("95% confidence interval:", round(t_test_result$conf.int[1], 2), "to", round(t_test_result$conf.int[2], 2), "\n")
```

```python
%%R
contingency_table <- table(df$service_type, df$is_delayed)
print("Contingency table (service_type rows, is_delayed columns):")
print(contingency_table)

chi_test <- chisq.test(contingency_table)
print(chi_test)
```

```python
%%R
library(ggplot2)
library(dplyr)

# Calculate delay rate and 95% CI for each service type
delay_props <- df %>%
  filter(!is.na(service_type), !is.na(is_delayed)) %>%
  group_by(service_type) %>%
  summarise(
    delay_rate = mean(is_delayed, na.rm = TRUE),
    n = n(),
    se = sqrt(delay_rate * (1 - delay_rate) / n),
    lower = delay_rate - 1.96 * se,
    upper = delay_rate + 1.96 * se
  )

# Bar chart with error bars
p <- ggplot(delay_props, aes(x = reorder(service_type, -delay_rate), y = delay_rate, fill = service_type)) +
  geom_col(width = 0.7) +
  geom_errorbar(aes(ymin = lower, ymax = upper), width = 0.2) +
  scale_y_continuous(labels = scales::percent, limits = c(0, 0.45)) +
  labs(
    title = "Delay Rate by Service Type (with 95% Confidence Interval)",
    subtitle = "No statistically significant association (χ² = 3.66, p = 0.454)",
    x = "Service Type", y = "Delay Rate (%)",
    caption = "Error bars = 95% confidence interval"
  ) +
  theme_minimal() +
  theme(legend.position = "none")

print(p)
```

##Section 2: Python Data Processing and Operational Analytics
###Markdown summary:
I reload the master CSV, perform extensive feature engineering (delay categories, route risk score, customer risk segmentation, hub efficiency score, etc.), then use NumPy and Pandas to compute descriptive statistics, zone summaries, hub rankings, customer risk segments, and fleet health.

```python
# Reload and re-parse dates
df = pd.read_csv("orders_full_final.csv")
for col in ["order_created_at", "dispatch_time", "delivery_completed_at"]:
    if col in df.columns:
        df[col] = pd.to_datetime(df[col], errors="coerce")

# ---- Feature engineering ----
# Actual delivery hours and delay
df["actual_delivery_hrs"] = (df["delivery_completed_at"] - df["dispatch_time"]).dt.total_seconds() / 3600
df["delay_hours"] = df["actual_delivery_hrs"] - df["promised_window_hours"]

# Delay categories
bins = [-np.inf, -1, 0, 2, 6, np.inf]
labels = ["Early", "On-Time", "Minor Delay", "Significant Delay", "Critical Delay"]
df["delay_category"] = pd.cut(df["delay_hours"], bins=bins, labels=labels)

# Binary flags
df["is_delayed"] = df["delivery_status"].isin(["Delayed", "Failed"]).astype(int)
df["is_failed"] = (df["delivery_status"] == "Failed").astype(int)
df["complaint_flag"] = (df["complaint_count"] >= 1).astype(int)

# Repeated high severity
high_sev_cust = df.groupby("customer_id")["any_high_severity"].sum().reset_index()
high_sev_cust.columns = ["customer_id", "total_high_sev"]
df = df.merge(high_sev_cust, on="customer_id", how="left")
df["repeated_high_severity"] = (df["total_high_sev"] >= 2).astype(int)

# Route risk score (min-max scaled)
def minmax(series):
    r = series.max() - series.min()
    return (series - series.min()) / r if r > 0 else series * 0

route_stats = df.groupby(["pickup_zone", "dropoff_zone"]).agg(
    route_delay_rate = ("is_delayed", "mean"),
    route_avg_override = ("manual_route_override_count", "mean"),
    route_avg_complaint = ("complaint_count", "mean")
).reset_index()
route_stats["route_risk_score"] = (
    0.40 * minmax(route_stats["route_delay_rate"]) +
    0.35 * minmax(route_stats["route_avg_override"]) +
    0.25 * minmax(route_stats["route_avg_complaint"])
)
df = df.merge(route_stats[["pickup_zone", "dropoff_zone", "route_risk_score"]],
              on=["pickup_zone", "dropoff_zone"], how="left")

# Driver override percentile
override_pct = df.groupby("driver_id")["manual_route_override_count"].mean()
df["driver_override_pctile"] = df["driver_id"].map(override_pct.rank(pct=True))

# Vehicle age
if "commission_date_x" in df.columns:
    df["commission_date"] = df["commission_date_x"]
if "commission_date" in df.columns:
    df["commission_date"] = pd.to_datetime(df["commission_date"], errors="coerce")
    df["vehicle_age_days"] = (pd.Timestamp("2025-01-01") - df["commission_date"]).dt.days
else:
    df["vehicle_age_days"] = np.nan

# In-repair flag
if "maintenance_status" in df.columns:
    df["maintenance_status"] = df["maintenance_status"].astype(str)
    df["in_repair_flag"] = (df["maintenance_status"] == "InRepair").astype(int)
else:
    df["in_repair_flag"] = np.nan

# Hub failure rate
hub_fail = df.groupby("hub_name")["is_delayed"].mean().rename("hub_failure_rate")
df = df.join(hub_fail, on="hub_name")

# Customer risk segment
df["customer_risk_segment"] = np.select(
    [
        (df["loyalty_score"] >= 7) & (df["complaint_flag"] == 1),
        (df["loyalty_score"] >= 7) & (df["complaint_flag"] == 0),
        (df["loyalty_score"] < 7) & (df["complaint_flag"] == 1),
    ],
    ["High-Value At-Risk", "High-Value Stable", "Low-Value Dissatisfied"],
    default="Low-Value Stable"
)

print("Feature engineering complete. New columns added:")
new_cols = ["delay_hours", "delay_category", "is_delayed", "is_failed", "complaint_flag",
            "repeated_high_severity", "route_risk_score", "driver_override_pctile",
            "vehicle_age_days", "hub_failure_rate", "customer_risk_segment", "in_repair_flag"]
print(new_cols)
print(df[new_cols].describe(include="all").T[["count", "mean", "std", "min", "max"]].to_string())
```

```python
new_features = [
    'delay_hours', 'delay_category', 'is_delayed', 'is_failed',
    'complaint_flag', 'repeated_high_severity', 'route_risk_score',
    'driver_override_pctile', 'vehicle_age_days', 'hub_failure_rate',
    'customer_risk_segment', 'in_repair_flag'
]

print("Missing features:", [c for c in new_features if c not in df.columns])

existing_features = [c for c in new_features if c in df.columns]
print("\nMissing values:")
print(df[existing_features].isna().sum().to_string())

print("\nBinary feature counts:")
for col in ['is_delayed', 'is_failed', 'complaint_flag', 'repeated_high_severity', 'in_repair_flag']:
    if col in df.columns:
        print(f"{col}: {df[col].value_counts(dropna=False).to_dict()}")

print("\nCategorical feature counts:")
for col in ['delay_category', 'customer_risk_segment']:
    if col in df.columns:
        print(f"{col}: {df[col].value_counts(dropna=False).to_dict()}")

print("\nKey numeric summaries:")
for col in ['vehicle_age_days', 'delay_hours', 'hub_failure_rate']:
    if col in df.columns:
        s = df[col]
        print(f"{col}: count={s.count()}, mean={s.mean():.3f}, min={s.min():.3f}, max={s.max():.3f}")
```

```python
# ---- NumPy descriptive statistics ----
num_cols = [
    "delay_hours", "order_value", "route_distance_km", "fuel_or_charge_cost",
    "manual_route_override_count", "battery_health_pct", "driver_rating",
    "training_score", "loyalty_score", "app_engagement_score"
]
arr = df[num_cols].dropna().to_numpy()
print("NumPy array shape:", arr.shape)
print("Column means  :", np.round(np.nanmean(arr, axis=0), 3))
print("Column medians:", np.round(np.median(arr, axis=0), 3))
print("Column std    :", np.round(np.nanstd(arr, axis=0), 3))
print("Column min    :", np.round(np.nanmin(arr, axis=0), 3))
print("Column max    :", np.round(np.nanmax(arr, axis=0), 3))

# ---- Zone-level operational summary ----
zone_summary = (
    df.groupby("pickup_zone")
    .agg(
        orders=("order_id", "count"),
        avg_delay_hrs=("delay_hours", "mean"),
        delay_rate=("is_delayed", "mean"),
        avg_overrides=("manual_route_override_count", "mean"),
        avg_order_value=("order_value", "mean"),
        total_revenue=("order_value", "sum"),
        total_fuel_cost=("fuel_or_charge_cost", "sum"),
        avg_complaints=("complaint_count", "mean")
    )
    .round(3)
    .sort_values("avg_delay_hrs", ascending=False)
    .reset_index()
)
print("\n--- Zone-level operational summary ---")
print(zone_summary.to_string(index=False))
```

```python
# ---- Hub efficiency score ----
hub_metrics = df.groupby('hub_name').agg(
    delay_rate=('is_delayed', 'mean'),
    overrides_rate=('manual_route_override_count', 'mean'),
    proof_missing_rate=('proof_of_completion_missing', 'mean')
).reset_index()
hub_metrics['efficiency_score'] = (
    0.40 * (1 - hub_metrics['delay_rate']) +
    0.35 * (1 - hub_metrics['overrides_rate']) +
    0.25 * (1 - hub_metrics['proof_missing_rate'])
)
print("\n--- Hub efficiency score ranking ---")
result = hub_metrics.sort_values('efficiency_score', ascending=False)[
    ['hub_name', 'efficiency_score', 'delay_rate', 'overrides_rate', 'proof_missing_rate']
].round(3)
print(result)
print(f"\nTop performer: {result.iloc[0]['hub_name']} (score: {result.iloc[0]['efficiency_score']:.3f})")
```

```python
# ---- Customer risk segment analysis ----
seg_analysis = (
    df.groupby("customer_risk_segment")
    .agg(
        customers=("customer_id", "nunique"),
        orders=("order_id", "count"),
        avg_loyalty=("loyalty_score", "mean"),
        avg_delay=("delay_hours", "mean"),
        avg_revenue=("order_value", "mean"),
        total_revenue=("order_value", "sum"),
        complaint_rate=("complaint_flag", "mean")
    )
    .round(3)
    .reset_index()
)
print("\n--- Customer risk segment analysis ---")
print(seg_analysis.to_string(index=False))
```

```python
# Build the document structure for each order
def build_service_case(row):
    return {
        "order_id": row["order_id"],
        "customer_id": row["customer_id"],
        "service_type": row.get("service_type"),
        "order_created_at": str(row.get("order_created_at", "")),
        "priority_level": row.get("priority_level"),
        "booking_channel": row.get("booking_channel"),
        "delivery": {
            "delivery_id": row.get("delivery_id"),
            "driver_id": row.get("driver_id"),
            "vehicle_id": row.get("vehicle_id"),
            "hub_id": row.get("hub_id"),
            "hub_name": row.get("hub_name"),
            "dispatch_time": str(row.get("dispatch_time", "")),
            "completed_at": str(row.get("delivery_completed_at", "")),
            "status": row.get("delivery_status"),
            "route_distance_km": row.get("route_distance_km"),
            "overrides": row.get("manual_route_override_count"),
            "rating": row.get("customer_rating_post_delivery"),
            "fuel_cost": row.get("fuel_or_charge_cost"),
            "proof_missing": bool(row.get("proof_of_completion_missing", 0))
        },
        "zone_context": {
            "pickup_zone": row.get("pickup_zone"),
            "dropoff_zone": row.get("dropoff_zone"),
            "home_zone": row.get("home_zone")
        },
        "complaints": {
            "count": int(row.get("complaint_count", 0)),
            "any_high_sev": bool(row.get("any_high_severity", 0)),
            "avg_compensation": row.get("avg_compensation"),
            "last_date": str(row.get("last_complaint_date", ""))
        },
        "app_events_summary": {
            "event_count": int(row.get("app_event_count", 0)),
            "unique_event_types": row.get("unique_event_types"),
            "avg_api_latency_ms": row.get("avg_api_latency"),
            "last_event": str(row.get("last_app_event", ""))
        },
        "risk_tags": {
            "is_delayed": bool(row.get("is_delayed", 0)),
            "complaint_flag": bool(row.get("complaint_flag", 0)),
            "route_risk_score": row.get("route_risk_score"),
            "delay_category": str(row.get("delay_category", ""))
        },
        "inserted_at": datetime.utcnow()
    }

# Insert all documents
collection = db["customer_cases"]
collection.drop()  # fresh start
service_docs = df.apply(build_service_case, axis=1).tolist()
result = collection.insert_many(service_docs)
print(f"Inserted {len(result.inserted_ids)} documents into 'customer_cases'")
```

```python
# ---- AGGREGATION: delay rate and compensation by zone ----
pipeline = [
    {"$group": {
        "_id": "$zone_context.pickup_zone",
        "total_orders": {"$sum": 1},
        "delayed_orders": {"$sum": {"$cond": ["$risk_tags.is_delayed", 1, 0]}},
        "avg_compensation": {"$avg": "$complaints.avg_compensation"},
        "total_complaints": {"$sum": "$complaints.count"},
        "avg_latency_ms": {"$avg": "$app_events_summary.avg_api_latency_ms"}
    }},
    {"$addFields": {"delay_rate": {"$divide": ["$delayed_orders", "$total_orders"]}}},
    {"$sort": {"delay_rate": -1}}
]
agg_results = list(collection.aggregate(pipeline))
print("\n--- Aggregation: Delay rate and avg compensation by zone ---")
for r in agg_results:
    print(r)
```

##Section: Query Optimisation and Performance Tuning
###Markdown summary:
I create indexes on frequently queried fields, compare explain() output before and after indexing, and document the performance improvement (COLLSCAN → IXSCAN).

```python
# Before indexing: explain plan
query = {
    "zone_context.pickup_zone": "Central",
    "risk_tags.is_delayed": True,
    "complaints.any_high_sev": True
}
print("--- BEFORE indexing ---")
explain_before = collection.find(query).explain()
print("Stage:", explain_before["queryPlanner"]["winningPlan"]["stage"])
if "executionStats" in explain_before:
    print("Docs examined:", explain_before["executionStats"]["totalDocsExamined"])
    print("Execution ms:", explain_before["executionStats"]["executionTimeMillis"])
else:
    print("(executionStats not available in free tier)")
```

```python
# Create indexes
print("\n--- Creating indexes ---")
collection.create_index(
    [("zone_context.pickup_zone", ASCENDING),
     ("risk_tags.is_delayed", ASCENDING),
     ("complaints.any_high_sev", ASCENDING)],
    name="zone_delay_complaint_idx"
)
collection.create_index([("delivery.driver_id", ASCENDING)], name="driver_id_idx")
collection.create_index([("delivery.hub_name", ASCENDING)], name="hub_name_idx")
collection.create_index([("order_id", ASCENDING)], unique=True, name="order_id_unique_idx")
collection.create_index([("inserted_at", DESCENDING)], name="inserted_at_desc_idx")
print("All indexes created.")

# List indexes
print("\nIndexes on customer_cases:")
for idx in collection.list_indexes():
    print(f"  {idx['name']} -> {idx['key']}")
```

```python
# After indexing: explain plan
print("\n--- AFTER indexing ---")
explain_after = collection.find(query).explain()
print("Stage:", explain_after["queryPlanner"]["winningPlan"]["stage"])
if "executionStats" in explain_after:
    print("Docs examined:", explain_after["executionStats"]["totalDocsExamined"])
    print("Execution ms:", explain_after["executionStats"]["executionTimeMillis"])
else:
    print("(executionStats not available)")

print("\nPerformance improvement: COLLSCAN -> IXSCAN, reduced examined documents.")
```

# MongoDB Analytical Insights

```python
# Insight 1: Hubs with most escalation flags
pipeline_hub_escalation = [
    {"$match": {"escalation_flag": "REVIEW_REQUIRED"}},
    {"$group": {
        "_id": "$delivery.hub_name",
        "count": {"$sum": 1},
        "avg_compensation": {"$avg": "$complaints.avg_compensation"},
        "avg_override_rate": {"$avg": "$delivery.overrides"},
        "sample_delay_categories": {"$push": "$risk_tags.delay_category"}
    }},
    {"$sort": {"count": -1}}
]

hub_escalation_results = list(collection.aggregate(pipeline_hub_escalation))

print("="*60)
print("Insight 1: Hubs with Most Escalation Flags (REVIEW_REQUIRED)")
print("="*60)
for r in hub_escalation_results:
    print(f"Hub: {r['_id']}")
    print(f"  Escalations: {r['count']}")
    print(f"  Avg Compensation: £{round(r.get('avg_compensation') or 0, 2)}")
    print(f"  Avg Override Rate: {round(r.get('avg_override_rate') or 0, 2)}")
    print(f"  Sample delay categories: {r['sample_delay_categories'][:3]}")
    print("-"*40)
```

```python
# Insight 2: App event latency by pickup zone
pipeline_latency_zone = [
    {"$match": {
        "app_events_summary.avg_api_latency_ms": {"$exists": True, "$ne": None}
    }},
    {"$group": {
        "_id": "$zone_context.pickup_zone",
        "avg_latency": {"$avg": "$app_events_summary.avg_api_latency_ms"},
        "max_latency": {"$max": "$app_events_summary.avg_api_latency_ms"},
        "orders": {"$sum": 1}
    }},
    {"$sort": {"avg_latency": -1}}
]

latency_results = list(collection.aggregate(pipeline_latency_zone))

print("="*60)
print("Insight 2: API Latency (ms) by Pickup Zone")
print("="*60)
for z in latency_results:
    print(f"Zone: {z['_id']}")
    print(f"  Avg latency: {round(z['avg_latency'], 1)} ms")
    print(f"  Max latency: {round(z['max_latency'], 1)} ms")
    print(f"  Orders with data: {z['orders']}")
    print("-"*40)
```

```python
# Insight 3: Cross‑tabulation of proof missing vs complaint flag
pipeline_proof_complaint = [
    {"$group": {
        "_id": {
            "proof_missing": "$delivery.proof_missing",
            "has_complaint": "$risk_tags.complaint_flag"
        },
        "count": {"$sum": 1},
        "avg_compensation": {"$avg": "$complaints.avg_compensation"},
        "total_compensation": {"$sum": "$complaints.avg_compensation"}
    }},
    {"$sort": {"_id.proof_missing": -1, "_id.has_complaint": -1}}
]

proof_complaint_results = list(collection.aggregate(pipeline_proof_complaint))

print("="*60)
print("Insight 3: Proof of Completion × Complaint Interaction")
print("="*60)
print(f"{'Proof Missing':<15} {'Has Complaint':<15} {'Count':<10} {'Avg Comp (£)':<15} {'Total Comp (£)'}")
print("-"*70)
for r in proof_complaint_results:
    proof = "Yes" if r['_id']['proof_missing'] else "No"
    complaint = "Yes" if r['_id']['has_complaint'] else "No"
    avg_comp = round(r.get('avg_compensation') or 0, 2)
    total_comp = round(r.get('total_compensation') or 0, 2)
    print(f"{proof:<15} {complaint:<15} {r['count']:<10} £{avg_comp:<14} £{total_comp}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Prepare data for bar chart
categories = []
counts = []
for r in proof_complaint_results:
    proof = "Missing" if r['_id']['proof_missing'] else "Has Proof"
    complaint = "+ Complaint" if r['_id']['has_complaint'] else "No Complaint"
    categories.append(f"{proof}\n{complaint}")
    counts.append(r['count'])

plt.figure(figsize=(8,5))
bars = plt.bar(categories, counts, color=['#d7301f' if 'Missing' in c else '#2c7bb6' for c in categories])
plt.title('Orders by Proof Status and Complaint Presence', fontsize=14)
plt.ylabel('Number of Orders')
plt.xlabel('Proof Status / Complaint')
for bar, count in zip(bars, counts):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 5, str(count), ha='center', va='bottom')
plt.tight_layout()
plt.show()
```
